```python
%%bash
cat > /tmp/test.cpp <<'CPP'
#include <iostream>
#include <string>

int countIndent(const std::string& line)
{
    std::string::const_iterator it = line.begin();
    int indent = 0;
    while (it != line.end())
    {
        if (*it == '\t')
        {
            ++indent;
            ++it;
        }
        else if (*it == ' ')
            return -1;
        else
            break;
    }
    return indent;
}

int main()
{
    const std::string line = "\t\tabc";
    std::cout << countIndent(line) << "\n";
}
CPP

g++ -std=c++98 -Wall -Wextra -Werror /tmp/test.cpp -o /tmp/test && /tmp/test
```
